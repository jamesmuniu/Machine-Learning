## Total Green House Gas Emissions as of 5th May 2022

```python
import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd
```

```python
greenhouse_emissions = pd.read_csv("world-bank-emissions.csv")
```

```python
greenhouse_emissions.head(5)
```

##### We will first make a copy of the original dataset in case we later want to make several different subsets, and for ease of use, we will rename one column into a shorter form at the same time:

```python
ge_data = greenhouse_emissions.copy().rename(columns={'Country Name':'Name'})
```

```python
ge_data.head(5)
```

##### Next, we will select the columns and rows we want to visualize in our chart. In this case, we will first choose the countries of Africa by name. A simple way of selecting a subset is by passing the loc method a set of conditions. The conditional expression is actually a pandas Series of boolean (True or False) values for each row in the original DataFrame. In this case, it returns True only for rows where the country name is in the list countries:

```python
countries =["Kenya", "South_Africa", "Egypt","Ghana","Nigeria","Morocco","Mauritius", "Mali"]
```

```python
ge_data= ge_data.loc[ge_data['Name'].isin(countries)]
```

```python
ge_data.head(7)
```

##### We will reformat the table for easier plotting, using the melt function to collect the relevant columns in a long format where each observation is a row, identified by the columns with the country name and the year. Since we only want the years 2000–2012, we take the columns from 34 onwards to value_vars—you can check what column name is at a particular index with ge_data.columns[34].



```python
ge_data_m = ge_data.melt(id_vars='Name', value_vars=ge_data.columns[14:], var_name='Year', value_name='Emissions')
```

```python
ge_data_m
```

#### Plotting the Data

##### Now that we have our data ready, we can begin plotting the line chart. One line of code gives a Seaborn line chart with default settings
##### sns.lineplot(data=ge_data_m, x='Year', y='Emissions', hue='Name')
##### The chart shows kt of CO2 equivalent of emissions by Ghana, Kenya, Morocco, Mali, Mauritius and Nigeria, . The horizontal axis ranges from 1990 to 2021 in increments of 1. The vertical axis ranges from 0 to 70000 in increments of 10000. The legend of the chart listing the name of the countries against the color of the line is in a box placed outside the chart. The approximate data from the chart are tabulated below.

```python
sns.set_theme(style="darkgrid")

ax = sns.lineplot(data=ge_data_m,x='Year', y='Emissions', hue='Name', palette="Dark2")

l =plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

fig = plt.gcf()

fig.set_size_inches(15,5)
```

##### In addition, we moved the legend outside the box by giving it some manual position attributes with plt.legend The given attributes put the upper left corner of the legend at 1, 1, that is, the top right corner of the chart area

```python
from matplotlib import ticker

  

ax =sns.lineplot(data=ge_data_m, x='Year', y='Emissions', hue='Name', palette="Dark2")

  

l =plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

fig = plt.gcf()

fig.set_size_inches(15,5)

  

fig.suptitle("Total greenhouse gas emissions, 2000–2018",size=16, weight='bold', ha='left', x=0.012)

fig.text(s='Source: World Bank, 2020', x=.9, y=0.05, ha='right', size=9)

plt.xlabel(None)

plt.ylabel("Emissions,kt of CO2 equivalent", rotation=0, position=(0,1.025), ha='left')

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
```

##### The somewhat cryptical command ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:0,.0f}")) sets y-axis ticks using a string method formatter, which allows extensive customization of how numbers are displayed. The value is represented by x and inputs after : determine formatting. Extra text can be put outside the brackets. To show the tick marks with two decimals and “kt” typed out instead you would replace the format string with "{x:,.2f}kt" See: https://pyformat.info/#number
##### Alternatively, we can add direct text labels for each line. This might be useful as the number of lines is at the upper limit of what can be distinguished by color code alone.




```python
from matplotlib import ticker

  

ax =sns.lineplot(data=ge_data_m, x='Year', y='Emissions', hue='Name', palette="Dark2")

  

l =plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

fig = plt.gcf()

fig.set_size_inches(15,7)

  

fig.suptitle("Total greenhouse gas emissions, 2000–2018",size=16, weight='bold', ha='left', x=0.012)

fig.text(s='Source: World Bank, 2020', x=.9, y=0.05, ha='right', size=9)

plt.xlabel(None)

plt.ylabel("Emissions,kt of CO2 equivalent", rotation=0, position=(0,1.025), ha='left')

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))


for row in ge_data_m[ge_data_m['Year']=='2018'].sort_values('Emissions').iterrows():

    plt.annotate(text= row[1].Name,xy=(row[1].Year, row[1].Emissions),

          textcoords='offset points',xytext=(20,0), arrowprops={'arrowstyle':'-', 'color':'gray'})

```

#### Because some of the series are overlapping, the labels end up overlapping as well, which is problematic. They could be repositioned in the code or manually adjusted after export.

#### We can also improve the appearance by calculating an alternative placement for the labels. One way to do that is by importing numpy and using the np.linspace function to create evenly spaced values over the value range of the last year in the series.

```python
import numpy as np
```

```python
ax = sns.lineplot(data=ge_data_m, x='Year', y='Emissions', hue='Name',

              palette="Dark2", lw=2)

  

l = plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

fig = plt.gcf()

fig.set_size_inches(15,7)

  

fig.suptitle("Total greenhouse gas emissions, 2000–2018", size=16, weight='bold', ha='left', x=0.012)

fig.text(s='Source: World Bank, 2020', x=.9, y=0.05, ha='right', size=9)

plt.xlabel(None)

plt.ylabel("Emissions,kt of CO2 equivalent", rotation=0, position=(0,1.025), ha='left')

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))

  

# fancy annotation

l.set_visible(False)# hide legend

  

colormap = {}

for line, text in zip(l.get_lines(), l.get_texts()):

    colormap[text.get_text()] =line.get_color()

  

last_values = ge_data_m[ge_data_m['Year']=='2018'].sort_values('Emissions').copy()

last_values['even']= np.linspace(last_values.Emissions.min(), last_values.Emissions.max(), len(last_values))

  

for row in last_values.iterrows():

    name = row[1].Name

    plt.annotate(text= name, xy=(row[1].Year,row[1].Emissions),

          weight='bold',

          color = colormap[name],

          xytext=(13,row[1].even),

          arrowprops={'arrowstyle':'-','color':'gray', 'relpos':(0,0.5)})
```
